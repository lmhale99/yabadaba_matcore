# MD test

Testing the MatCore-MD specification for the iprPy isolated_atom calculation.

## Intial steps

```python
import datetime

import yabadaba
import yabadaba_matcore
```

Specify a local XML database for storing the MatCore record.

```python
database = yabadaba.load_database(style='local', host='../matcore_database', format='xml')
```

## Create data transformer and define missing values

```python
matcore = yabadaba.load_record('matcore_md')

# Basic info
matcore.add_creator(name='Lucas Hale', affiliation='NIST')
matcore.title = f'Isolated atom potential energies'
matcore.creation_date = datetime.date.today()
matcore.description = 'Single non-interacting atom energies computed for an interatomic potential.  Used for relating potential energies to cohesive energies.'
matcore.matcore_id = 'iprPy-calculation_isolated_atom'
matcore.matcore_date = datetime.date.today()
matcore.license = 'NIST-Software'

# Material info
matcore.add_material(
    phase = 'Cluster',
    description = 'Single atom',
    constituent = dict(species='Any', concentration=100),
)

# Computation info
matcore.add_computation(
    method_class = 'Atomistic',
    method = 'MD',
    simulation_conditions = dict(
        type = 'Nonstandard',
        number_of_particles = 1,
    ),
    software = dict(
        name = 'LAMMPS'
    ),
    custom = 'tubular'
)

# MD info
matcore.add_md_computation(mode = 'static')
matcore.particle_style = 'Atom'
matcore.add_particle_interactions(
    model_type = 'Any',
    bonding_type = 'Reactive',
    theory_level = 'Classical physics-based',
    source = dict(
        reference = f'NIST IPR'
    )
)

matcore.name = matcore.matcore_id
```

```python
print(matcore.build_model().xml(indent=2))
```

```python
database.add_record(record=matcore)
```
